```python
import sys
import pyspark
from pyspark.sql import SparkSession
```

```python
SPARK_MASTER_HOST = "spark-master"
SPARK_MASTER_PORT = 7077
UPPER_BOUND = 1000 
```

```python
sys.version
```

```python
pyspark.__version__
```

```python
spark = SparkSession.builder  \
    .appName("ExampleGrandataApp")  \
    .master(f"spark://{SPARK_MASTER_HOST}:{SPARK_MASTER_PORT}")  \
    .getOrCreate()
```

```python
# Detalles de la sesion
spark
```

```python
my_range = spark.range(UPPER_BOUND).toDF("number")
divisBy5 = my_range.where("number % 5 = 0")
print(f"Total number of multiples of 5 from 0 to {UPPER_BOUND}: {divisBy5.count()}")
```

```python
spark.stop()
```
